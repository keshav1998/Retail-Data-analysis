```python
import pandas as pd
import numpy as np
```

```python
d1 = {'track_no':pd.Series([2,4,7,9,11,13,17]),
      'cars':pd.Series(['Mustang', 'Perati', 'Mitsubishi', 'Bugatti', 'Lamborghini','Ferrari', 'Tesla'])
     }
df1 = pd.DataFrame(d1)
```

```python
df1
```

```python
d2 = {'track_no':pd.Series([3,4,8,9,15,17]),
      'color':pd.Series(['yellow', 'red', 'blue', 'purple', 'olive','black'])
     }
df2 = pd.DataFrame(d2)
```

```python
df2
```

```python
car_info = df1.merge(df2, on='track_no', how = 'inner')
car_info
```

```python
car_info = df1.merge(df2, on='track_no', how = 'outer')
car_info
```

```python
car_info = df1.merge(df2, on='track_no', how = 'left')
car_info
```

```python
car_info = df1.merge(df2, on='track_no', how = 'right')
car_info
```
